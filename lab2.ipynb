# 1.Matplotlib Website

    This chart shows of creating a stacked bar plot using bar chart.

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Snares\n $\\mu=$3700.66g",
    "Emperor\n $\\mu=$3733.09g",
    "Aptenodytes\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

# 2. Seaborn Website
    This graph shows the Timeseries plot with error bands

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

# 3. Plotly Express
     This chart shows Strip Charts with Plotly Express
     

```python
import plotly.express as px
import plotly.offline as offline
import plotly.graph_objs as go

offline.init_notebook_mode()
df = px.data.tips()
fig = px.strip(df, x="total_bill", y="time", orientation="h", color="smoker")
fig.show()
```
